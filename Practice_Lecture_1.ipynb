### `Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
def convert_data_type(series,data_type):
    return pd.to_numeric(series,errors='coerce',downcast='float')
````

````python
series = pd.Series([1,2,'Python',2.0,True,100])
````

````python
convert_data_type(series,'float')
````

### `Q-9`
    i. Read `items.csv` making `item_name` as index.
    ii. Show no of nan values
    ii. Item price is given in $, so convert it to rupees without currency symbol.
    iii. Make data type of newly made series as float.
    iv. Fill nan with mean of the series


How csv file looks

```
item_name	item_price
Chips and Fresh Tomato Salsa	$2.39
Izze	$3.39
Nantucket Nectar	$3.39
Chips and Tomatillo-Green Chili Salsa	$2.39
Chicken Bowl	$16.98

```

````python
item = pd.read_csv('items.csv')
item.head()
````

````python
item.set_index('item_name',inplace=False)
````

````python
item.isnull().sum()
````

````python
item['item_price'] = item['item_price'].astype(str).str.replace('$', '', regex=False).astype(float)
````

````python
item
````

````python
item['item_price'] = item['item_price'].astype(str).str.split('$').apply(lambda x:float(x[0])*88.74)
item
````

````python
item['item_price'].fillna(item['item_price'].mean())
````

````python
item.set_index('item_name')
````
