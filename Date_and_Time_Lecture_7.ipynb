## Timestamp Objects
> Time stamps reference particular moments in time (e.g.,Oct 24th,2022 at 7:00pm)

### Creating timestamp objects

```python
## variations
pd.Timestamp('2025/10/13')
pd.Timestamp('2025-10-13')
pd.Timestamp('2025.10.13')
# pd.Timestamp('2025,10,13')  ## not work
pd.Timestamp('2025, 10, 13')
```

```python
## only year
pd.Timestamp('2025')
```

```python
## using text
pd.Timestamp('13th Oct 2025')
```

```python
## providing time also
pd.Timestamp('2025, 10, 13, 1:13pm')
pd.Timestamp('2025, 10, 13, 1:13')
```

```python
## using datetime.datetime object
import datetime as dt
dt.datetime(2025,10,13,1,19,50)
```

```python
x = pd.Timestamp(dt.datetime(2025,10,13,1,19,50))
x
```

```python
## fetching attributes
x.year
x.month
x.day
x.hour
x.minute
x.second
```

```python
## why seprate objects to handle date and time when python already has datetime functionality
```

* Syntax wise datetime is very convenient
* but the performance takes a hit while working with huge data . List vs Numpy Array
* The weekness of Python's datetime format inspired the Numpy team to add a set of native time series data type to Numpy.
* The datetime64  dtype encodes dates as 64-bit integers , and thus allows arrays of dates to be represented very compactly

```python
import numpy as np
date = np.array('2025-08-04',dtype=np.datetime64)
date
```

```python
date + np.arange(12)
```

* Because of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large
* Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64
* From a group of these Timestamp objects, Pandas can construct a Datetimelndex that can be used to index data in a Series or DataFrame

### date_range function

```python
## generate daily dates in a given range
pd.date_range(start='2025/10/01',end='2025/10/30',freq='D')
```

```python
pd.date_range(start='2025/10/01',end='2025/10/30',freq='2D')
```

```python
pd.date_range(start='2025/10/01',end='2025/10/30',freq='3D')
```

```python
## B -> business days
pd.date_range(start='2025/10/01',end='2025/10/30',freq='B')
```

```python
## W -> one week per day
pd.date_range(start='2025/10/01',end='2025/10/30',freq='W')
```

```python
pd.date_range(start='2025/10/01',end='2025/10/30',freq='W-Thu')
```

```python
## H -> Hourly date(factor)
pd.date_range(start='2025/10/01',end='2025/10/30',freq='h')
```

```python
pd.date_range(start='2025/10/01',end='2025/10/30',freq='6h')
```

```python
## ME -> month end
pd.date_range(start='2025/10/01',end='2025/12/05',freq='ME')
```

```python
## MS ->month start
pd.date_range(start='2025/10/01',end='2025/12/05',freq='MS')
```

```python
## YE -> year end
pd.date_range(start='2025/10/01',end='2029/12/05',freq='YE')
```

```python
## YS -> year start
pd.date_range(start='2025/10/01',end='2029/12/05',freq='YS')
```

```python
## using periods (number of results)
pd.date_range(start='2025/10/13',periods=25,freq='5h')
```

### to_datetime function
>  converts an existing objects to pandas timestamp / datetimeindex object

```python
## simple series example

s = pd.Series(['2025/10/13','2024/10/13','2023/10/13'])
pd.to_datetime(s).dt.year
pd.to_datetime(s).dt.month
pd.to_datetime(s).dt.day
pd.to_datetime(s).dt.month_name()
pd.to_datetime(s).dt.day_name()
```

```python
## with errors

s = pd.Series(['2025/10/13','2024/10/13','2023/100/13'])
print(s)
pd.to_datetime(s,errors='coerce')
```
