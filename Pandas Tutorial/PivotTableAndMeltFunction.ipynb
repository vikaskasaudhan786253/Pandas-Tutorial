Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,20,13,14]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["days"])
```

```python
pd.melt(var,id_vars=["days"],var_name="Python")
```

```python
pd.melt(var,id_vars=["days"],var_name="Python",value_name="wscube")
```

Pivot()

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":['a','b','b','a','b','b'],"eng":[10,12,14,15,16,12],"maths":[17,18,19,20,13,14]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

```python
var = pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":['a','b','b','a','b','b'],"eng":[10,12,14,15,16,12],"maths":[17,18,19,20,13,14]})
var
```

```python
var.pivot_table(index="st_name",columns='days',aggfunc="mean",margins = True)
```
