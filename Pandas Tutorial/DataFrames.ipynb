```python
import pandas as pd
```

```python
l = [1,2,3,4,5,6]
var = pd.DataFrame(l)
print(var)
print(type(var))
```

```python
d = {"a":[1,2,3,4,5],"s":[4,3,2,1,5],"d":[9,6,7,4,5],"f":[4,31,21,22,5]}
var = pd.DataFrame(d)
print(var)
print(type(var))
print(var["a"][3])
```

```python
d = {"a":[1,2,3,4,5],"s":[4,3,2,1,5],"d":[9,6,7,4,5],"f":[4,31,21,22,5]}
var = pd.DataFrame(d,columns=[])
print(var)
```
