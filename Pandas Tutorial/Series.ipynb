SERIES

```python
import pandas as pd
```

```python
x = [1,3,5,7,9,6,8,2,4]
var = pd.Series(x)
print(var)
print(type(var))
print(var[5])
```

index changing

```python
x = [1,3,5,7]
var = pd.Series(x,index=['a','s','d','f'])
print(var)
```

```python
x = [1,3,5,7]
var = pd.Series(x,index=['a','s','d','f'],dtype='float')
print(var)
```

```python
x = [1,3,5,7]
var = pd.Series(x,index=['a','s','d','f'],dtype='float',name="Python")
print(var)
```

```python
dic = {"name":['Python','C','C++','Java'],"por":[12,13,14,15],"rank":[1,4,3,2]}

var = pd.Series(dic)
print(var)
```

```python
s = pd.Series(12)
print(s)
```

```python
s = pd.Series(12,index=[1,2,3,4,5,6])
print(s)
print(type(s))
```

```python
s1 = pd.Series(12,index=[1,2,3,4,5,6])
s2 = pd.Series(12,index=[3,4,5,6])

print(s1+s2)
```
