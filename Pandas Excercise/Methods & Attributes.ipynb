# 2. Indexing & Selection

## a. .loc[label], .iloc[pos], .at[label, col], .iat[pos, col] — row/column access



```python
df.head(3)
```

### Label-based Access

```python
df.loc[1,'Ticket']
```

```python
df.loc[[0,1,2],'Ticket']
```

```python
df.loc[[0,111,2],['Name','Sex','Age']]
```

### Integer Position-based Access

```python
df.iloc[1,0]
```

```python
df.iloc[1,[0,1,2,3,4]]
```

```python
df.iloc[[0,1,2,3,4],3]
```

```python
df.iloc[[0,1,2,3],[2,3,4]]
```

### Fast Label-based Scalar Access

```python
df.at[1,'Sex']
```

```python
df.at[8,'Sex']
```

### Fast Integer-position Scalar Access

```python
df.iat[2,5]
```

## b. .get(...), .reindex() — safe lookup & alignment



```python
s = pd.Series({'a':10,'b':20,'c':30,'d':40})
print(s.get('b'))
print(s.get('x'))
print(s.get('x',0))
```

```python
data = pd.DataFrame({'Name':['A','B','C'],'Score':[50,60,70]},index=[0,1,2])
new_data=data.reindex([2,0,3])
new_data
```

## c. .filter(...), .query(expr) — selection by labels or boolean expression



### ✅ 1. .filter() — Select rows/columns by labels or regex


```python
df.filter(items=['Sex','Age'],axis=1)
```

```python
df.filter(items=[1,2,5,30],axis=0)
```

```python
df.filter(like='Ti',axis=1)
```

```python
df.filter(regex='^C',axis=1)
```

### ✅ 2. .query() — Filter rows using a boolean expression (SQL-like)


```python
df.query('Age>50')
```

```python
df.query('Age>50 and Sex == "female"')
```

## d. .isin(), .between() — containment testing



### ✅ 1. .isin() — Check Membership (Like in for lists)


```python
data = pd.DataFrame(df)
```

```python
data
```

```python
data['Sex'].isin(['male','female'])
```

### ✅ 2. .between() — Check if Values Fall in a Range


```python
data['Age'].between(30,50,inclusive='both')
```

## e. .take(),.slice_shift(),.swaplevel(), .reorder_levels()---advanced indexing

### 1. .take()---select rows/column by integer position

```python
df.take([1,4,7,9],axis=0)
```

```python
df.take([2,4,6,8,9],axis=1)
```

### .slice_shift()---shift rows while keeping the same size

```python
data = pd.DataFrame(df)
```

```python
data.shift(periods=1)
```

```python
data.shift(periods=-1)
```

### .swaplevel()---swp levels in multiIndex

```python
data.swaplevel(i=4,j=5,axis=0)
```
