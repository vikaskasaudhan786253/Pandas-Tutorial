#### Series from lists
 '''copy = True''' will not change in original array and list but '''copy = False''' will change in original array but not in list

```python
## string
country = ['India','Pakistan','Nepal',"Shrilanka",'USA']
pd.Series(country)
```

```python
## Integers
runs = [31,23,45,67,43,87]
run_series = pd.Series(runs)
run_series
```

```python
runs = [31,23,45,67,43,87]
run_series = pd.Series(runs,copy=False)
run_series.iloc[0] = 234
print(run_series)
print(runs)
```

```python
runs = [31,23,45,67,43,87]
run_series = pd.Series(runs,copy=True)
run_series.iloc[0] = 234
print(run_series)
print(runs)
```

```python
## custom index  (default index is replaced by subjects)
marks = [45,78,98,99]
subjects = ['English','Hindi','Science','Math']
pd.Series(marks,index=subjects)
```

```python
## setting a name
marks = pd.Series(marks,index=subjects,name='Vikas_ka_marks')
marks
```

### Series from dictionary

```python
marks = {
        'math':99,
        'english':67,
        'science':88,
        'hindi':88
}
marks_series = pd.Series(marks,name='vikas_ka_marks')
marks_series
```

```python
marks_series = pd.Series(marks,name='vikas_ka_marks',index=[1,2,3,4])
marks_series
```

### series from 1D array

```python
arr = np.array([1,2,3,4,5,6,7])
arr
```

```python
ser = pd.Series(arr,copy=False)
ser.iloc[0]=345
print(ser)
print(arr)
```

```python
brr = np.array([1,2,3,4,5,6,7])
ser = pd.Series(brr,copy=True)
ser.iloc[0]=35
print(ser)
print(brr)
```
